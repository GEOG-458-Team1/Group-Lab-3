## Part 1 - Data Description & Overview

The following data includes a geographic shapefile with respective estimates data, provided by the Washington State’s Office of Financial Management. The census data is measured at the block level from 2010, as part of the Office of Financial Management’s *Small area estimates program*. Using the census block data, estimates for population and housing are made throughout Washington State. After processing the estimates data and shapefile, the data will be categorized into “urban” and “rural”.

According to the [Wikipedia page](https://en.wikipedia.org/wiki/Urban_area), the U.S. Census Bureau defines an urban area as:

> Census blocks that have a population density of at least 1,000 people per square mile.

> Respective surrounding census blocks that have an overall density of at least 500 people per square mile. 

As for rural areas, the U.S. Census Bureau’s definition is:

> Open country and settlements with less than 2,500 residents, or a max population density of 999 people per square mile.

For our purposes, we will define "rural" as areas with a population density of 999 people per square mile or less.

However, the U.S. Census Bureau's definition of urban is not universal. According to the UK Census, the United Kingdom uses the term "built-up area" instead of the term "urban area".
According to the [Office of National Statistics](https://en.wikipedia.org/wiki/Office_for_National_Statistics), the term "built-up area" means:
> areas with built-up land with a minimum area of 20 hectares (200,000m<sup>2</sup>), while settlements within 200 meters of each other are linked. Built-up area sub-divisions are also identified to provide greater detail in the data, especially in the large conurbanations. 

Another difference in the definition of urban can also be seen in the Philippines. The cities in the Philippines are governed by the [Local Government Code of 1991](https://www.lawphil.net/statutes/repacts/ra1991/ra_7160_1991.html) and the [municipal charter](https://en.wikipedia.org/wiki/Municipal_charter) which states that highly urbanized cities are:
> local government units autonomous from provinces that have a minimum population of 200,000 and an annual income of at least 50 million.

Lastly, according to the [Canada 2011 Census](https://en.wikipedia.org/wiki/2011_Canadian_Census), the term "urban areas" was changed to "population centre". This is due to the belief that there isn't a strict division between the definition of urban and rural but instead there are multiple settlement patterns that exist. Although there is a change in terminology, the definition of population centre remains the same as:
>A population of at least 1,000 people where the density is no fewer than 400 persons per km<sup>2</sup>

A population centre is then further classified into three categories:
> small (population 1,000 to 29,999), medium (population 30,000 to 99,999) and large (population 100,000 or greater)

While many different places use different terms to represent "urban", the definition of most of them have to do with population. This can be seen in the U.S, Canada, as well as the Philippines. However, according to the UK Census, the term "built-up area" seems to focus on the area of land instead of the population, which presents an interesting comparison in the view of the term "urban area".

Here I read in the DBF file that contains the County Name and the FIPS County. This FIPS County is what we will use to kind of work with the shapefile contents since **'FIPSCounty'** is present also in the shapefile content. In the shapefile, this data is called **'COUNTYFP10'**.

```python
fp2 = "data\WashingtonFIPS.dbf"
db = ps.open(fp2) # Pysal to open DBF
dictionary = {col: db.by_col(col) for col in db.header} #convert dbf to dictionary
fips = pd.DataFrame(dictionary) #convert dictionary to pandas
display(fips)
```

This is the part where I create the loop that will create the shapefiles for each county that contain the block groups of that county. You can read more on how I did this in the comments below!

```python
#Create a dictionary from the DBS file 
#Create list from panda table
codes = fips["FIPSCounty"].tolist()
names = fips["CountyName"].tolist()

#Convert unicode strings in lists to regular strings
codes = [x.encode('UTF8') for x in codes]
names = [x.encode('UTF8') for x in names]

#Create dictionary from the two list using the FIPS as the key
d = dict(zip(codes,names))

#Group the shapefile by the COUNTYFP10 
grouped = data.groupby('COUNTYFP10')

#Output path name
folder = r"data\county_shapefiles"

#Create shapefiles by County by grouping the shapefile with 'COUNTYFP10'
for key, values in grouped: 
        #The data of that county with each block group data
        county = values 
        #Creating the name of the output file by using the dictionary that we 
        #created early to get the name of the county 
        outname = ("%s.shp" % d[key].replace(" ", "_")) 
        #Printing where the process is in the loop
        print("Processing: County %s" % key)
        #Writing the file to the folder
        outpath = os.path.join(folder,outname)
        values.to_file(outpath)


```

This is where we create the ranked list of population by county. Using the metadata of the shapefile to get the Population data of 2018. The column name in the shapefile that represents this data is **'POP2018'**.  You can see more details below on how I create the ranked list below.

```python
#Create empty lists to store data for county names and population sums
county2 = []
population = []

#Go through each group and sum the population for each group
for key, values in grouped: 
    county = values
    county2.append(d[key])
    population.append(int(county['POP2018'].sum()))
    

populationDF = pd.DataFrame(list(zip(county2,population)),
                           columns=['County', 'Population'])

#Display top 10 Counties with the most Population
populationDF.sort_values(by=["Population"], ascending=False).head(10)
```
